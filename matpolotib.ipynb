```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]   # some random values

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Line 1 data
x1 = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]

# Line 2 data
x2 = [1, 2, 3, 4, 5]
y2 = [2, 3, 5, 7, 11]

# Plotting two lines
plt.plot(x1, y1, label="Squares")      # First line
plt.plot(x2, y2, label="Primes")       # Second line

plt.title("Two Lines in One Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.legend()   # show labels
plt.grid(True)

plt.show()
```
