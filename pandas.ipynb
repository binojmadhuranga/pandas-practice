```python
import pandas as pd

# Sample data
data = {
    "Name": ["Ann", "John", "Sam"],
    "Age": [20, 22, 25],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

# 1. Shape
print("Shape of the DataFrame:")
print(df.shape)

# 2. Info
print("\nInfo of the DataFrame:")
print(df.info())

# 3. Describe
print("\nDescribe (Statistics):")
print(df.describe())
```
